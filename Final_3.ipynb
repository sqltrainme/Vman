```python
import pandas as pd

import numpy as np

import matplotlib

import seaborn as sns
%matplotlib inline

df = pd.read_csv('https://data.cityofnewyork.us/resource/qgea-i56i.csv',index_col=0)
df

df.drop(columns=['cmplnt_fr_dt', 'cmplnt_fr_tm','cmplnt_to_dt','cmplnt_to_tm','addr_pct_cd','ky_cd','pd_cd','crm_atpt_cptd_cd','jurisdiction_code','parks_nm','hadevelopt','housing_psa','x_coord_cd','y_coord_cd','transit_district','latitude','longitude','station_name'], axis=1)



df = pd.read_csv('https://data.cityofnewyork.us/resource/qgea-i56i.csv',usecols=['ky_cd'])
print(df.describe())

sns.histplot(df,x='ky_cd')
```
